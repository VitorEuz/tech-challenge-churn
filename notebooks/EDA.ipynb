```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# ======================
# 1. Carregar dados
# ======================
df = pd.read_csv("../data/raw/telco_customer_churn.csv")

print("Primeiras linhas:")
display(df.head())

print("\nShape do dataset:")
print(df.shape)

print("\nInformações gerais:")
df.info()

# ======================
# 2. Limpeza inicial
# ======================
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"] = df["TotalCharges"].fillna(0)

print("\nValores nulos após limpeza:")
print(df.isna().sum())

# ======================
# 3. Distribuição do Churn
# ======================
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Churn", hue="Churn", palette="viridis")
plt.title("Distribuição da Classe Churn")
plt.xlabel("Churn")
plt.ylabel("Quantidade de Clientes")
plt.show()

print("\nPercentual de Churn:")
print(df["Churn"].value_counts(normalize=True) * 100)

# ======================
# 4. Tempo de contrato vs Churn
# ======================
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="tenure", hue="Churn", multiple="stack", palette="viridis")
plt.title("Tempo de Contrato (tenure) vs Churn")
plt.xlabel("Meses de Contrato")
plt.ylabel("Quantidade de Clientes")
plt.show()

# ======================
# 5. Cobrança mensal vs Churn
# ======================
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Churn", y="MonthlyCharges")
plt.title("Monthly Charges vs Churn")
plt.xlabel("Churn")
plt.ylabel("Cobrança Mensal")
plt.show()

# ======================
# 6. Tipo de contrato vs Churn
# ======================
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Contract", hue="Churn")
plt.title("Tipo de Contrato vs Churn")
plt.xlabel("Tipo de Contrato")
plt.ylabel("Quantidade de Clientes")
plt.xticks(rotation=20)
plt.show()

# ======================
# 7. Método de pagamento vs Churn
# ======================
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="PaymentMethod", hue="Churn")
plt.title("Método de Pagamento vs Churn")
plt.xlabel("Método de Pagamento")
plt.ylabel("Quantidade de Clientes")
plt.xticks(rotation=30)
plt.show()

# ======================
# 8. Serviços de internet vs Churn
# ======================
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="InternetService", hue="Churn")
plt.title("Serviço de Internet vs Churn")
plt.xlabel("Serviço de Internet")
plt.ylabel("Quantidade de Clientes")
plt.show()

# ======================
# 9. Insights finais
# ======================
print("""
INSIGHTS PRINCIPAIS:

1. A base está desbalanceada: há mais clientes No Churn do que Churn.
2. Clientes com menor tenure tendem a cancelar mais.
3. Clientes com contrato Month-to-month apresentam maior risco de churn.
4. Clientes com cobrança mensal mais alta aparecem mais associados ao churn.
5. O método Electronic check apresenta maior associação com churn.
6. Esses padrões justificam o uso de modelos de classificação para prever risco de churn.
""")
```
